```python
import pandas as pd
```

```python
data_df = pd.read_csv("./data/annotations/phase1/train/train.csv")
data_df = data_df.rename(columns={"# Age-Young": "Age-Young"})
```

```python
age_counts = data_df[[c for c in data_df.columns if c.startswith("Age")]].sum()
```

```python
age_counts
```

```python
import matplotlib.pyplot as plt

plt.pie(age_counts.values, labels=age_counts.index, autopct='%1.1f%%', shadow=True, startangle=90)

# Add a title
plt.title('Age')

# Show the chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```

```python
gender_counts = {"Male": len(data_df) - data_df["Gender-Female"].sum(), "Female": data_df["Gender-Female"].sum()}
```

```python
plt.pie(gender_counts.values(), labels=gender_counts.keys(), autopct='%1.1f%%', shadow=True, startangle=90)

# Add a title
plt.title('Gender')

# Show the chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```

```python
ub_color_counts = data_df[[c for c in data_df.columns if c.startswith("Hair-Length")]].sum()
```

```python
plt.pie(ub_color_counts.values, labels=[c.split("-")[-1] for c in ub_color_counts.index], autopct='%1.1f%%', shadow=True, startangle=90)

# Add a title
plt.title('Hair Length')

# Show the chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```

```python
ub_length_counts = {"Long": len(data_df) - data_df["UpperBody-Length-Short"].sum(), "Short": data_df["UpperBody-Length-Short"].sum()}
```

```python
plt.pie(ub_length_counts.values(), labels=ub_length_counts.keys(), autopct='%1.1f%%', shadow=True, startangle=90)

# Add a title
plt.title('Upper Body Length')

# Show the chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```
